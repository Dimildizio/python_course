## 1. Работа с файлами: создать, дозаписать, прочитать

У функции `open` есть важный параметр - режим:

- `"w"` — **запись**. Создаёт новый файл или ПОЛНОСТЬЮ затирает старый.
- `"a"` — **дозапись**. Добавляет текст в конец файла.
- `"r"` — **чтение**.

Создадим файл `battle_log.txt` и запишем туда две строки (разделяя их ```\n```).


**Вопрос для самоизучения** (часто на собесах услышать можно) - что делает (кроме открытия файла) with зачем он нужен и как работает

Как это работает:

1. Сперва мы используем контекстный менеджер with и "открываем" нужный нам ресурс (в данном случае файл) с open. (Есть и более длинный подход через try/finally и .close())

2. С помощью open() создаем объект файла, указывая имя файла и режим, а также давая имя переменной объекту файла с помощью ```as имя```

3. У такого объекта есть метод write который мы и используем для того чтоб записать наши данные в файл
